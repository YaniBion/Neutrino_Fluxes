```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%config InlineBackend.figure_format = 'retina'
import math
from matplotlib import rc


def plotsdefault(xlab='',ylab='',\
                 lw=1,lfs=12,tfs=12,dpi=300,size_x=5,size_y=3,Grid=False):
    plt.rcParams.update(plt.rcParamsDefault)

    plt.rcParams['axes.linewidth'] = lw
#     plt.rc('text', usetex=True)
    matplotlib.rcParams["figure.dpi"] = dpi
    matplotlib.rc('font', family='sans-serif') 
    matplotlib.rc('font', serif='Helvetica Neue') 

    fig = plt.figure(figsize=(size_x,size_y))
    ax = fig.add_subplot(111)

    ax.set_xlabel(xlab,fontsize=lfs)
    ax.set_ylabel(ylab,fontsize=lfs)

#     ax.tick_params(which='major',direction='in',width=1,length=13,right=True,top=True,pad=7)
#     ax.tick_params(which='minor',direction='in',width=1,length=10,right=True,top=True)
    if Grid:
        ax.grid()
    return fig,ax

def find_nearest_return_approximation(array,array2,value):
    # Find nearest value in array and returns array2 evaluated in that index
    idx = np.searchsorted(array, value, side="left")
    if idx > 0 and (idx == len(array) or math.fabs(value - array[idx-1]) < math.fabs(value - array[idx])):
        return array2[idx-1]
    else:
        return array2[idx]
    
```

```python
folder = './neutrino_data/'
paths ={ 'atmospheric': 'Atmospheric.dat', 'sun_thermal' : 'Sun-thermal.dat', 
        'sun_pp' : 'Sun-nuclear-pp.dat', 'sun_O15' : 'Sun-nuclear-O15.dat', 'sun_N13' : 'Sun-nuclear-N13.dat',
       'sun_hep' : 'Sun-nuclear-hep.dat', 'sun_B8' : 'Sun-nuclear-B8.dat', 'sun_lines' : 'Sun-lines.dat',
       'reactor' : 'Reactor.dat', 'icecube' : 'IceCube.dat', 'geo' : 'Geoneutrinos.dat', 'DSNB' : 'DSNB.dat',
       'cosmo' : 'Cosmogenic.dat', 'CNB' : 'CNB.dat', 'CNB_lines' : 'CNB-lines.dat', 'BBN_n' : 'BBN-neutron.dat',
       'BBN_H3': 'BBN-tritium.dat'}
```

```python
energy_atmospheric = []
atmospheric_neutrinos = []
atmospheric_antineutrinos = []

with open(folder+paths.get('atmospheric')) as file:
    lines = file.readlines()
    for line in lines[4:]:
        e = float(line.split('   ')[1])
        f1 = float(line.split('   ')[2])
        f2 = float(line.split('   ')[3])
        energy_atmospheric.append(e)
        atmospheric_neutrinos.append(f1)
        atmospheric_antineutrinos.append(f2)

energy_thermal = []
thermal = []
with open(folder+paths.get('sun_thermal')) as file:
    lines = file.readlines()
    for line in lines[4:]:
        e = float(line.split('   ')[1])
        f1 = float(line.split('   ')[2])
        energy_thermal.append(e)
        thermal.append(f1)
        
energy_pp = []
pp = []
with open(folder+paths.get('sun_pp')) as file:
    lines = file.readlines()
    for line in lines[4:]:
        e = float(line.split('   ')[1])
        f1 = float(line.split('   ')[2])
        energy_pp.append(e)
        pp.append(f1)
        
energy_O15 = []
O15 = []
with open(folder+paths.get('sun_O15')) as file:
    lines = file.readlines()
    for line in lines[5:]:
        e = float(line.split('   ')[1])
        f1 = float(line.split('   ')[2])
        energy_O15.append(e)
        O15.append(f1)
        
energy_N13 = []
N13 = []
with open(folder+paths.get('sun_N13')) as file:
    lines = file.readlines()
    for line in lines[4:]:
        e = float(line.split('   ')[1])
        f1 = float(line.split('   ')[2])
        energy_N13.append(e)
        N13.append(f1)
        
energy_hep = []
hep_min = []
hep_max = []

with open(folder+paths.get('sun_hep')) as file:
    lines = file.readlines()
    for line in lines[4:]:
        e = float(line.split('   ')[1])
        f1 = float(line.split('   ')[2])
        f2 = float(line.split('   ')[3])
        energy_hep.append(e)
        hep_min.append(f1)
        hep_max.append(f2)

energy_B8 = []
B8 = []

with open(folder+paths.get('sun_B8')) as file:
    lines = file.readlines()
    for line in lines[4:]:
        e = float(line.split('   ')[1])
        f1 = float(line.split('   ')[2])
        energy_B8.append(e)
        B8.append(f1)
        
energy_sunlines = []
sunlines = []

with open(folder+paths.get('sun_lines')) as file:
    lines = file.readlines()
    for line in lines[4:]:
        e = float(line.split('   ')[1])
        f1 = float(line.split('   ')[2])
        energy_sunlines.append(e)
        sunlines.append(f1)
        
energy_reactor = []
reactor = []     

with open(folder+paths.get('reactor')) as file:
    lines = file.readlines()
    for line in lines[4:]:
        e = float(line.split('   ')[1])
        f1 = float(line.split('   ')[2])
        energy_reactor.append(e)
        reactor.append(f1)        

energy_icecube = []
icecube = []     

with open(folder+paths.get('icecube')) as file:
    lines = file.readlines()
    for line in lines[4:]:
        e = float(line.split('   ')[1])
        f1 = float(line.split('   ')[2])
        energy_icecube.append(e)
        icecube.append(f1)  
        
        
energy_geo = []
geo = []     

with open(folder+paths.get('geo')) as file:
    lines = file.readlines()
    for line in lines[4:]:
        e = float(line.split('   ')[1])
        f1 = float(line.split('   ')[2])
        energy_geo.append(e)
        geo.append(f1)  
        
energy_DSNB = []
nu_DSNB_min = []     
nu_DSNB_max = []     
antinu_DSNB_min = []     
antinu_DSNB_max = []     

with open(folder+paths.get('DSNB')) as file:
    lines = file.readlines()
    for line in lines[4:]:
        e = float(line.split('   ')[1])
        f1 = float(line.split('   ')[2])
        f2 = float(line.split('   ')[3])
        af1 = float(line.split('   ')[4])
        af2 = float(line.split('   ')[5])

        energy_DSNB.append(e)
        nu_DSNB_min.append(f1)
        nu_DSNB_max.append(f2)
        antinu_DSNB_min.append(af1)  
        antinu_DSNB_max.append(af2)  

energy_cosmo = []
cosmo_He = []     
cosmo_Fe = []     

with open(folder+paths.get('cosmo')) as file:
    lines = file.readlines()
    for line in lines[4:]:
        e = float(line.split('   ')[1])
        f1 = float(line.split('   ')[2])
        f2 = float(line.split('   ')[3])

        energy_cosmo.append(e)
        cosmo_He.append(f1)  
        cosmo_Fe.append(f2)  
        
energy_CNB = []
CNB = []     

with open(folder+paths.get('CNB')) as file:
    lines = file.readlines()
    for line in lines[4:]:
        e = float(line.split('   ')[1])
        f1 = float(line.split('   ')[2])
        energy_cosmo.append(e)
        cosmo_He.append(f1)  
    
energy_CNB_lines = []
CNB_lines = []     

with open(folder+paths.get('CNB_lines')) as file:
    lines = file.readlines()
    for line in lines[4:]:
        e = float(line.split('   ')[1])
        f1 = float(line.split('   ')[2])
        energy_CNB_lines.append(e)
        CNB_lines.append(f1)  
    
energy_BBN_n = []
BBN_n = []     

with open(folder+paths.get('BBN_n')) as file:
    lines = file.readlines()
    for line in lines[4:]:
        e = float(line.split('   ')[1])
        f1 = float(line.split('   ')[2])
        energy_BBN_n.append(e)
        BBN_n.append(f1)  

    
energy_BBN_H3 = []
BBN_H3 = []     

with open(folder+paths.get('BBN_H3')) as file:
    lines = file.readlines()
    for line in lines[4:]:
        e = float(line.split('  ')[1])
        f1 = float(line.split('   ')[2])
        energy_BBN_H3.append(e)
        BBN_H3.append(f1)  
        
x = np.logspace(-1, 0.5, int(1e5))
x2 = np.logspace(0.5, 3, 55)
x = np.append(x,x2)
S = []
epsilon = (max(x)-min(x))/len(x)

for value in x:
    s = np.zeros(10)
    if value < max(np.asarray(energy_pp)/1e6) and value >  min(np.asarray(energy_pp)/1e6):
        s[0] = find_nearest_return_approximation( np.asarray(energy_pp)/1e6, np.asarray(pp)*1e6,value)
    if value < max(np.asarray(energy_O15)/1e6)and value >  min(np.asarray(energy_O15)/1e6):
        s[1] = find_nearest_return_approximation( np.asarray(energy_O15)/1e6,np.asarray(O15)*1e6, value)
    if value < max(np.asarray(energy_N13)/1e6)and value >  min(np.asarray(energy_N13)/1e6):
        s[2] = find_nearest_return_approximation( np.asarray(energy_N13)/1e6, np.asarray(N13)*1e6,value)
    if value < max(np.asarray(energy_B8)/1e6)and value >  min(np.asarray(energy_B8)/1e6):
        s[3] = find_nearest_return_approximation( np.asarray(energy_B8)/1e6, np.asarray(B8)*1e6,value)
    if value < max(np.asarray(energy_hep)/1e6)and value >  min(np.asarray(energy_pp)/1e6):
        s[4] = find_nearest_return_approximation( np.asarray(energy_hep)/1e6, (np.asarray(hep_min)+np.asarray(hep_max))/2*1e6,value)
    if value < max(np.asarray(energy_atmospheric)/1e6)and value >  min(np.asarray(energy_pp)/1e6):
        s[5] = find_nearest_return_approximation( np.asarray(energy_atmospheric)/1e6,(np.asarray(atmospheric_neutrinos) + np.asarray(atmospheric_antineutrinos))*1e6, value)
    if value < max(np.asarray(energy_DSNB)/1e6) and value >  min(np.asarray(energy_DSNB)/1e6):
        s[6] = find_nearest_return_approximation( np.asarray(energy_DSNB)/1e6,(np.asarray(nu_DSNB_min) + np.asarray(antinu_DSNB_min)+np.asarray(nu_DSNB_max) + np.asarray(antinu_DSNB_max))/2*1e6,value)
    if value > energy_sunlines[0]/1e6 - epsilon and  value < energy_sunlines[0]/1e6 + epsilon:
        s[7] = sunlines[0]*1e6
    if value > energy_sunlines[1]/1e6 - epsilon and  value < energy_sunlines[1]/1e6 + epsilon:
        s[8] = sunlines[1]*1e6
    if value > energy_sunlines[2]/1e6 - epsilon and  value < energy_sunlines[2]/1e6 + epsilon:
        s[9] = sunlines[2]*1e6
        
    S.append(sum(s))
```

```python
colors = ['firebrick', 'gray', 'navy', 'lime', 'lightskyblue', 'slategray', 'salmon', 'dodgerblue'
         , 'gold', 'crimson']
plotsdefault(r'E$_\nu$[keV]', r'Neutrino Flux [cm$^{-2}$ keV$^{-1}$ s$^{-1}$]')

plt.plot(np.asarray(energy_pp)/1e6,np.asarray(pp)*1e6, color=colors[0], label='pp')
plt.plot(np.asarray(energy_O15)/1e6,np.asarray(O15)*1e6,color= colors[1], label=r'$^{15}$O')
plt.plot(np.asarray(energy_N13)/1e6,np.asarray(N13)*1e6, color=colors[2], label=r'$^{13}$N')
plt.plot(np.asarray(energy_B8)/1e6,np.asarray(B8)*1e6,color= colors[3], label=r'$^8$B')
plt.vlines(energy_sunlines[0]/1e6,1e-4, sunlines[0]*1e6,color= colors[7], label=r'$^7$Be1')
plt.vlines(energy_sunlines[1]/1e6,1e-4, sunlines[1]*1e6,color= colors[8], label=r'$^7$Be2')
plt.vlines(energy_sunlines[2]/1e6,1e-4, sunlines[2]*1e6, color=colors[9], label=r'pep')
plt.plot(np.asarray(energy_hep)/1e6,(np.asarray(hep_min)+np.asarray(hep_max))/2*1e6, color=colors[4], label=r'hep')
plt.fill_between(np.asarray(energy_hep)/1e6,np.asarray(hep_min)*1e6,np.asarray(hep_max)*1e6, 
                 alpha=0.3, color=colors[4])

plt.plot(np.asarray(energy_atmospheric)/1e6,(np.asarray(atmospheric_neutrinos) + np.asarray(atmospheric_antineutrinos))*1e6,
         color=colors[5], label=r'Atmospheric')
plt.fill_between(np.asarray(energy_DSNB)/1e6,(np.asarray(nu_DSNB_min) + np.asarray(antinu_DSNB_min))*1e6,
                (np.asarray(nu_DSNB_max) + np.asarray(antinu_DSNB_max))*1e6, 
                 alpha = 0.3,color=colors[5])
plt.plot(np.asarray(energy_DSNB)/1e6,(np.asarray(nu_DSNB_min) + np.asarray(antinu_DSNB_min)+np.asarray(nu_DSNB_max) + np.asarray(antinu_DSNB_max))/2*1e6,
         color=colors[6], label=r'DSNB')


plt.plot(x,S, '--',lw=0.8,color='black')

plt.xscale('log')
plt.yscale('log')
plt.xlim(9e-2,1e3)
plt.ylim(1e-4, 1e12)
plt.legend(frameon=False,prop={'size': 8})
plt.show()
```
